## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

Ranking

We are not just classifying good/bad pages. We need to order search results by relevance for each keyword.
Ranking models learn to put the best pages at the top based on multiple signals like backlinks, CTR, and content quality.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

Answer: position_in_serp and click_through_rate

The label comes from observed outcome. We get this from Google Search Console.
For each keyword-url pair we record the actual ranking position and if users clicked.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

Answer: NDCG@10 and CTR@1

We measure if the best pages are at the top 10. NDCG checks ranking quality.
CTR@1 checks if the first result actually gets clicks.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

Answer: backlinks, content_score, page_speed, ctr

We will use: number of referring domains, content relevance score, page load speed, and historical click through rate.
These features predict if a page deserves to rank high.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

Answer: position_bias and clickbait

Risk: People click top results just because they’re first, not because they’re best.
Guardrail: Compare CTR to average CTR for that position. Also check bounce rate to avoid clickbait.
